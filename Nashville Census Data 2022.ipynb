```python
nash_poverty_cd_2022 = nash_cd_2022[['GISJOIN', 'YEAR', 'STATE', 'COUNTY', 'TRACTA', 'GEOID', 'AMZME001', 'AMZME002', 'AMZME003']]
```

```python
nash_poverty_cd_2022.rename(columns = {'AMZME001' : 'TotalHouseholds', 'AMZME002' : 'UnderHalfPL', 'AMZME003' : 'UnderPL'}, inplace = True)
```
