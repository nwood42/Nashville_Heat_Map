```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from tabulate import tabulate
```

```python
cd_1988 = pd.read_csv("C:/Data/Nash_Census_2022/nhgis0018_csv/nhgis0018_csv/nhgis0018_ds123_1990_tract.csv")
```

```python
tn_cd_1988 = cd_1988[(cd_1988['STATE'] == 'Tennessee')]
```

```python
nash_cd_1988 = tn_cd_1988[(tn_cd_1988['COUNTY'] == 'Davidson')]
```

```python
nash_poverty_cd_1988 = nash_cd_1988[['GISJOIN', 'YEAR', 'STATE', 'COUNTY', 'TRACTA', 'E1C001', 'E1C002', 'E1C003', 'E1C004', 'E1C005', 'E1C006', 'E1C007', 'E1C008', 'E1C009']]
```

```python
nash_poverty_cd_1988.rename(columns = {'E1C001' : 'UnderHalfPL', 'E1C002' : 'UnderPL'}, inplace = True)
```

```python
nash_poverty_cd_1988['PercentUnderPL'] = ((nash_poverty_cd_1988.UnderHalfPL + nash_poverty_cd_1988.UnderPL) /
(nash_poverty_cd_1988.UnderHalfPL + nash_poverty_cd_1988.UnderPL + nash_poverty_cd_1988.E1C003 + nash_poverty_cd_1988.E1C004 + nash_poverty_cd_1988.E1C005 + nash_poverty_cd_1988.E1C006 + nash_poverty_cd_1988.E1C007 + nash_poverty_cd_1988.E1C008 + nash_poverty_cd_1988.E1C009)) * 100
```
